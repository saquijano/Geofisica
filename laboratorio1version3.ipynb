```python
x=linspace(-2000,2000,4001)
```

```python
c= 6.674*10**(-11)
s=2400
pi=np.pi
```

```python
def diapiros(z,d,r):
    g=4/3*(d-s)*pi*c*(r**3)*((z/(x**2+z**2)**(3/2)))
    return g
```

```python
plot(x,diapiros(-200, 2200, 60),label="200")
plot(x,diapiros(-400, 2200, 60),label="400")
plot(x,diapiros(-600, 2200, 60),label="600")
plt.legend(loc=0)
plt.xlabel("Metros")
plt.ylabel("Anomalias gravimetricas (miligal)")
plt.title("Anomalias gravimetricas causadas por un diapiro enterrado a 3 profundidades")
```

```python
c= 6.674*10**(-11)
s=2400
def losas(z,p,h):
    l=2*c*h*(p-s)*(pi/2+(np.arctan(x/z)))
    return l
```

```python
plot(x,losas(-200, 2200, 60),label="200")
plot(x,losas(-400, 2200, 60),label="400")
plot(x,losas(-600, 2200, 60),label="600")
plt.legend(loc=0)
plt.xlabel("Metros")
plt.ylabel("Anomalias gravimetricas (miligal)")
plt.title("Anomalias gravimetricas causadas por placa semi infinita a profundidades")
```
